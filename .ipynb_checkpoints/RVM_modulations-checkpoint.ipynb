```python
gs_top = plt.GridSpec(3, 1, hspace=0.0,height_ratios=[1.2,1.2, 2])
gs_base = plt.GridSpec(3, 1,height_ratios=[1.2,1.2, 1.4],hspace=0.3)


fig = plt.figure(figsize=(10,10))


topax = fig.add_subplot(gs_top[0,:])
other_axes = [fig.add_subplot(gs_top[i,:], sharex=topax) for i in range(1,2 )]
top_axes = [topax] + other_axes
sequence=np.array([1/6,1/5,1/4,1/3,1/2,1])
label=[r'$t=T_{\rm eff}/6$',r'$t=T_{\rm eff}/5$',r'$t=T_{\rm eff}/4$', r'$t=T_{\rm eff}/3$',r'$t=T_{\rm eff}/2$',\
       r'$t=T_{\rm eff}$']

data=np.genfromtxt('biaxial_alpha1.dat')
t=data[:,0]
a=data[:,1]
alphat=sp_interp1d(t, a)
period=np.array([5.2068*1e7, 5.90265*1e7,7.51673*1e7,8.68127*1e7])

def rvm(alpha,i,Phi):
    Psi=np.arctan(sin(alpha)*sin(Phi)/ (-sin(i)*cos(alpha)+cos(i)*sin(alpha)*cos(Phi)))
    return Psi

for j in range(6):
    alpha=alphat(sequence[j]*period[0])

    i=radian(45)
    Phi=np.linspace(-radian(30),radian(30),100)
    s1=rvm(alpha,i,Phi)
    top_axes[0].plot(Phi/pi*180,s1/pi*180,linewidth=2,color=q(j/6),label=label[j])
    
top_axes[0].legend(fontsize=22,ncol=3,loc=(0.0,1.05)) 


data=np.genfromtxt('triaxial_alpha1.dat')
t=data[:,0]
a=data[:,1]
alphat=sp_interp1d(t, a)
period=np.array([5.2068*1e7, 5.90265*1e7,7.51673*1e7,8.68127*1e7])

def rvm(alpha,i,Phi):
    Psi=np.arctan(sin(alpha)*sin(Phi)/ (-sin(i)*cos(alpha)+cos(i)*sin(alpha)*cos(Phi)))
    return Psi

for j in range(6):
    alpha=alphat(sequence[j]*period[2])
    print(alpha/pi*180)
    i=radian(45)
    Phi=np.linspace(-radian(30),radian(30),100)
    s1=rvm(alpha,i,Phi)
    top_axes[1].plot(Phi/pi*180,s1/pi*180,linewidth=2,color=q(j/6))
    
top_axes[1].set_xlim(-20,20)
top_axes[0].minorticks_on()
top_axes[1].minorticks_on()
axs[2].minorticks_on()
axs[0].set_ylim(-92,92)
axs[1].set_ylim(-92,92)

top_axes[0].set_ylabel(r'$\Psi\,(\rm deg)$',fontsize=25)
top_axes[1].set_ylabel(r'$\Psi\,(\rm deg)$',fontsize=25)
top_axes[1].set_xlabel(r'$\Phi\,(\rm deg)$',fontsize=25)

for ax in top_axes[:-1]:
    plt.setp(ax.get_xticklabels(), visible=False)

bottomax = fig.add_subplot(gs_base[2,:]) 

data=np.genfromtxt('biaxial_alpha1.dat')
t=data[:,0]
a=data[:,1]
alphat=sp_interp1d(t, a)
period=np.array([5.2068*1e7, 5.90265*1e7,7.51673*1e7,8.68127*1e7])

t1=np.linspace(0,3*period[0],200)
alpha=alphat(t1)
beta=radian(45)-alpha

bottomax.plot(t1/yr, -sin(beta)/sin(alpha),linewidth=2,label=r'$\rm biaxial$')
bottomax.set_xlabel(r'$t\,(\rm yr)$',fontsize=30)
bottomax.set_ylabel(r'$-\sin \beta/\sin \alpha$',fontsize=30)
# bottomax.set_xlim(0,1.4)
# bottomax.set_ylim(-0.20,1.8)


data=np.genfromtxt('triaxial_alpha1.dat')
t=data[:,0]
a=data[:,1]
alphat=sp_interp1d(t, a)
period=np.array([5.2068*1e7, 5.90265*1e7,7.51673*1e7,8.68127*1e7])

t1=np.linspace(0,3*period[2],200)
alpha=alphat(t1)
beta=radian(45)-alpha
bottomax.plot(t1/yr, -sin(beta)/sin(alpha),linewidth=2,color='r',linestyle='dashed',label=r'$\rm triaxial$')

bottomax.set_xlim(0,5)
bottomax.set_ylim(-0.95,0.6)
bottomax.minorticks_on()
bottomax.legend(fontsize=25,frameon=False,ncol=2,loc=(0.1,0.02))
plt.text(2.2,2.9,r'$\rm biaxial$',fontsize=25)
plt.text(2.2,1.5,r'$\rm triaxial$',fontsize=25)

plt.savefig("../../paper/fig_rvm1.pdf", format='pdf', bbox_inches="tight")
```
